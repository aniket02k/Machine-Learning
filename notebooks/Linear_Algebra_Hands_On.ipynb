# **1. Scalar, Vectors, Matrix and Tensor in NumPy**

```python
#Scalar = single number
scalar=1.0
print(scalar)
print("Type of scalar", type(scalar))

print("-"*40)

#Vector -> 1D data
vector = np.array(
    [3,5,7]
)
print(vector)
print("Shape of Vector", vector.shape)

print("-"*40)
#Matrix data

matrix = np.array([[1,2,3],
                  [4,5,6]
                ])

print(matrix)
print("Shape of Matrix", matrix.shape)
print("-"*40)

tensor=np.array(
    [
        [
            [1,2,3],
            [4,5,6]
        ],
        [
            [11,12,13],
            [14,15,16]
        ]
    ]
)

print(tensor)
print("Shape of Matrix", tensor.shape)
print("-"*40)
```

### **2. Creating Vectors and Matrices for ML Style Data**

```python
#Imagine we have 3 samples and 2 features
# Feature1: hours studied
# Feature2: Number of questions solved
# target - Passed(1)/Fail(0)


X=np.array(
    [
        [2,30],   #Sample1
        [5,80],   #Sample2
        [1,20]    #Sample3
    ]
)


y=np.array(
    [0,1,0]
)

print("Input Data Matrix X:\n",X)
print("X Shape", X.shape) #(n_samples, n_features)

print("Output Vector: y:\n",y)
print("Y Shape", y.shape)


#A single sample as a vector
x1=X[0]

print("\n First sample as a vector:",x1)
print("\nx1 Shape:",x1.shape)
```

# **3. Basic Operations: Addition, Scalar Multiplication, Transpose**

```python
#Create two vectors with same length
v1= np.array([1,2,3])
v2= np.array([4,5,6])

print("v1:",v1)
print("v2:",v2)

v_add=v1+v2

print("Vector addition v1+v2:",v_add)


#Scalar Multiplication

v1_scaled= 2 * v1
print("Vector addition 2 * v1:",v1_scaled)

#Matrics Example

A = np.array(
    [
      [1,2],
      [3,4]
    ]
)
B = np.array(
    [
      [10,20],
      [30,40]
    ]
)


print("Matrix A:\n",A)
print("Matrix B:\n",B)

#Matrix addition
print("Matrix Addition A + B:\n", A+B)

#Scalar Multiplication

print("Scalar Multiplication 0.5*A\n",0.5 *A)

#Matrix Transpose

print("Transpose of A:\n", A.T)

```
